```python
# This program creates an application that interacts with a webservice to obtain data when a user enters either city or
# zip code via a request from weather forecast data http://openweathermap.org/. Once the data is obtained, the weather
# forecast is displayed in a readable format to the user. Further, the program uses comments to specify its courses,
# uses functions, allows the user to run it multiple times in order to obtain weather forecast for multiple locations,
# checks the accuracy of the data the user enters, notifies the user in the event that the data entered is incorrect,
# and uses the request library to gain data from the webservice.
# Name Frenel Cherubin
# Date 6/01/2019
#Course DSC 510
# Assignment number 12
import requests  # The "requests" resource facilitate the interpretor(Python) along with the IDE(Pycharm)to gain infor-
# mation from a web service.


def usWeatherStd(x):
# Obtaining the URL! The URL contains the API key
    url = "http://api.openweathermap.org/data/2.5/weather?q=" + x + ",us&APPID=b9594f5dfdbf1c20b4c963e01f09ecb8&units=imperial"
    response = requests.request("GET", url)
    weather = response.json()
    t = (weather['main']['temp'])      # t represents temperature
    h = (weather['main']['humidity'])  # h is the humidity
    d = float(weather['wind']['deg'])  # d represents degree
    s = (weather['wind']['speed'])     # s is the speed
    n = (weather['name'])
    print("The current temp in ", n, "is ", t, ' Degrees Fahrenheit')
    print('The humidity is at ', h, '%.')
    print('The wind is blowing at ', s, ' KPH')
    if d > 0 and d <= 90:
        print('Out of the southwest')
    elif d > 90 and d <= 180:
        print('Out of the southeast')
    elif d > 180 and d <= 270:
        print('Out of the northeast')
    elif d > 270 and d <= 360:
        print('Out of the northwest')
    else:
        print("The direction where the weather is coming from is unknown. " )
# Obtaining the URL! The URL contains the API key

def osWeather(x):
    url = "http://api.openweathermap.org/data/2.5/weather?q=" + x + "&APPID=b9594f5dfdbf1c20b4c963e01f09ecb8&units=metric"
    response = requests.request("GET", url)
    weather = response.json()
    t = (weather['main']['temp'])     # t represents temperature
    h = (weather['main']['humidity']) # h is the humidity
    d = float(weather['wind']['deg']) # d represents degree
    s = (weather['wind']['speed'])    # s is the speed
    n = (weather['name'])
    print("The current temp in ", n, "is ", t, " degrees Celsius.")
    print('The humidity is at ', h, '%')
    print('The wind is blowing at ', s, ' KPH')
    if d > 0 and d <= 90:
        print('Out of the southwest')
    elif d > 90 and d <= 180:
        print('Out of the southeast')
    elif d > 180 and d <= 270:
        print('Out of the northeast')
    elif d > 270 and d <= 360:
        print('Out of the northwest')
    else:
        print("The direction where the weather is coming from is unknown. " )


def main():
    choosing = True
    while choosing == True: #This section of the program facilitates the user to run it several times to loop up weather.
        type_choice = str(input('Please enter the name of the city or zip code to obtain the weather forecast \n'
                        'enter 1 for National, 2 for International. '
                        '(Any other key to exit) : ')) # This asks the user for input
# In this section, try blocks are used to make the request successful.
        try:
            if type_choice == '1':
                x = input("Please enter the name of the city or zip code to obtain the weather forecast : ")
                usWeatherStd(x)
            elif type_choice == '2':
                x = input("Please enter the name or zip code of the foreign city you want to obtain the weather forecast : ")
                osWeather(x)
            else: choosing = False
            print("Thanks for using our weather service request library. Be safe!")

        except:
            print('Input is invalid, try again.')
main()
```
